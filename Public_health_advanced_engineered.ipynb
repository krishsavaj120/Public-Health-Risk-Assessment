```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('diabetes_binary_health_indicators_BRFSS2015.csv')
```

```python
df.head(10)
```

```python
df.columns = [col.lower().replace(' ', '_') for col in df.columns]
```

```python
df.head(10)
```

```python
age_map = {1:'18-24', 2:'25-29', 3:'30-34', 4:'35-39', 5:'40-44', 6:'45-49', 
           7:'50-54', 8:'55-59', 9:'60-64', 10:'65-69', 11:'70-74', 12:'75-79', 13:'80+'}
df['age_bracket'] = df['age'].map(age_map)
```

```python
df.head(10)
```

```python
def classify_bmi(bmi):
    if bmi < 18.5: return 'Underweight'
    elif bmi < 25: return 'Normal'
    elif bmi < 30: return 'Overweight'
    else: return 'Obese'
df['bmi_category'] = df['bmi'].apply(classify_bmi)
```

```python
df.head(10)
```

```python
df['risk_score'] = (df['highbp']*2 + df['highchol']*2 + (df['bmi']>=30)*3 + df['heartdiseaseorattack']*3 + df['smoker']*1)
```

```python
df.head(10)
```

```python
income_map = {1: '<$10k', 2: '$10k-$15k', 3: '$15k-$20k', 4: '$20k-$25k', 
              5: '$25k-$35k', 6: '$35k-$50k', 7: '$50k-$75k', 8: '>$75k'}
df['income_bracket'] = df['income'].map(income_map)
```

```python
gen_hlth_map = {1: 'Excellent', 2: 'Very Good', 3: 'Good', 4: 'Fair', 5: 'Poor'}
df['health_perception'] = df['genhlth'].map(gen_hlth_map)
```

```python
df['lifestyle_score'] = df['physactivity'] + df['fruits'] + df['veggies']
```

```python
df['vulnerability_idx'] = df['stroke'] + df['heartdiseaseorattack'] + df['diffwalk']
```

```python
def bin_mental_health(days):
    if days == 0: return 'Stable'
    elif days <= 5: return 'Occasional Stress'
    elif days <= 15: return 'Frequent Stress'
    else: return 'High Distress'
df['mental_health_status'] = df['menthlth'].apply(bin_mental_health)
```

```python
df['access_barriers'] = (1 - df['anyhealthcare']) + df['nodocbccost']
```

```python
df.head(10)
```

```python
df.to_csv('public_health_advanced_engineered.csv', index=False)
```
